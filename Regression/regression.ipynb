**Q1)** Create a regression model between TotalWorkingYears and MonthlyIncome
(response variable). Show the scatter plot with regression line between them.


```python
model = sm.OLS(df["MonthlyIncome"], sm.add_constant(df["TotalWorkingYears"])).fit()
model.summary()
```

```python
plt.plot(df['TotalWorkingYears'].values, df['MonthlyIncome'].values,'ro')
plt.plot(df['TotalWorkingYears'].values ,model.fittedvalues.values,  'b')
plt.legend(['Data', 'Fitted model'])
plt.xlabel('TotalWorkingYears')
plt.ylabel('MonthlyIncome')
```

###Intrepretation of the given result:
- R-squared value indicates that approximately 59.7% of the variance in MonthlyIncome is explained by the variation in TotalWorkingYears. The closer R-squared is to 1, the better the model fits the data, so 0.597 suggests a reasonable fit. <br>
- p value is less than 0.05 which indicates that we can reject our null hypothesis, in this case which is that there is no correlation between the two factors, hence suggesting that there is a significant relationship between TotalWorkingYears and MonthlyIncome. <br>
- Constant Coefficient is the estimated value of MonthlyIncome when TotalWorkingYears is zero, in this case which is 1227.9353, approximately 1228. Coefficient is 467.6584 â‰ˆ 467 that suggests that for a one unit increase in TotalWorkingYears, the MonthlyIncome increases by 467 units.

**Q4)** Create a regression model to predict MonthlyIncome using all other inputs. Discuss
the effectiveness of the model. Report 3 most significant inputs and 3 least significant
inputs (based on p-value) and interpret the results. Create a new input and show the
prediction of MonthlyIncome using the same model.

First we will find the columns with categorical data as regression does not work on them,
and using one-hot encoding to convert them to numerical data.

```python
col = df.select_dtypes(exclude = ['int', 'float']).columns.tolist()
col
```

```python
df1 = pd.get_dummies(df, columns = [i for i in col])
X = df1.drop(["MonthlyIncome"], axis = 1)
Y = df1["MonthlyIncome"]
model = sm.OLS(Y, sm.add_constant(X)).fit()
model.summary()
```

###Intrepretation of the above results
- Given the R-squared value of 0.944, the model seems to have large proportion of variance in the data. The model seems to be quite effective and caters almost 94.4% of variability in the data.
- Based on the p-values of different variables, the most siginificant variables are  
    - JobLevel
    - TotalWorkingYears
    - JobRoles

 They all have p-value less than 0.05, including almost all the JobRoles, hence they are third most significant input.

 The three most following least significant input based on p-value is;
 - YearsInCurrentRole
 - PerformanceRating
 - Department

 They all have p-value greater than 0.05, including almost all the departments, hence they are the third least significant input.


```python
new = pd.DataFrame({
    'Age': [23],
    'Department': ['Research & Development'],
    'DistanceFromHome': [20],
    'Education': [4],
    'EducationField': ['Life Sciences'],
    'Gender': ['Male'],
    'HourlyRate': [80],
    'JobLevel': [3],
    'JobRole': ['Research Scientist'],
    'MaritalStatus': ['Single'],
    'NumCompaniesWorked': [2],
    'OverTime': ['No'],
    'PerformanceRating': [3],
    'StockOptionLevel': [0],
    'TotalWorkingYears': [2],
    'TrainingTimesLastYear': [3],
    'WorkLifeBalance': [3],
    'YearsAtCompany': [2],
    'YearsInCurrentRole': [2],
    'YearsSinceLastPromotion': [0],
    'YearsWithCurrManager': [2]
    })

new = pd.concat([df, new], ignore_index=True)
new = new.drop(['MonthlyIncome'], axis=1)
df2 = pd.get_dummies(new, columns=[i for i in col])
df2 = sm.add_constant(df2)
new = df2.tail(1)
model.predict(new)
```

Using this model, the predicted MonthlyIncome is 7915.
