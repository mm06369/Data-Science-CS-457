```python
data = pd.read_csv('formated.csv')
```

Fetching columns that have null values


```python
data.columns[data.isna().any()]
```

Dropping those two columns because it is completely null.


---



```python
data.drop(columns=['Access Code', 'Email Address'], inplace=True)
```

Taking mode of values in column of 'Region', and placing that value in any empty cell.

```python
repeated_region = data["Region"].mode()
data["Region"].fillna(repeated_region.values[0], inplace=True)
col = data.columns[data.isna().any()].tolist()
data['Region']
```

Taking mode of values in 'City/ Town/ District' column, and placing the mode value in any empty cell.


```python
repeated_region = data["City / Town / District"].mode()
data["City / Town / District"].fillna(repeated_region.values[0], inplace=True)
col = data.columns[data.isna().any()].tolist()
data["City / Town / District"]
```

One of the age value is 1993 which is year. So making sure if year is entered subtract if from 2023 and enter the age


```python
import numpy as np


data['Q3'] = data['Q3'].str.replace(' ', '')
data["Q3"] = data['Q3'].astype(int)
data['Q3'] = np.where(data['Q3'] > 100, 2023 - data['Q3'], data['Q3'])
data['Q3']
```

In Q4 we have height in inches. We will take the mean of all the height data, and fill the column with null value with the mean value. Also I will correct some anomalies in the data like 6l, and 7O.

```python
data['Q4'] = np.where(data['Q4'] == "7O", "70", data['Q4'])
data['Q4'] = np.where(data['Q4'] == "6l", "61", data['Q4'])
data['Q4'] = data['Q4'].astype(float)

mean = data["Q4"].mean()
data["Q4"].fillna(mean, inplace=True)
data['Q4']
```

Taking mode for values in Q6, and assigning it to null values.

```python
mode_value = data["Q5"].mode()
data["Q5"].fillna(mode_value.values[0], inplace=True)
data['Q5'] = data['Q5'].str.strip()
data['Q5'] = data['Q5'].str.upper()
data['Q5'] = np.where(data['Q5'] == 'UNITED STATES ; PAKISTAN',
                     'UNITED STATES', data['Q5'])

data['Q5']
```

```python
mode_value = data["Q6"].mode()
data["Q6"].fillna(repeated_region.values[0], inplace=True)
data['Q6'] = data['Q6'].str.strip()
data['Q6'] = data['Q6'].str.upper()
data['Q6'] = data['Q6'].replace('YES',np.NaN)
data["Q6"].fillna(mode_value.values[0], inplace=True)
data['Q6']
```

```python
repeated_region = data["Q7"].mode()
data["Q7"].fillna(repeated_region.values[0], inplace=True)
data['Q7']
```

In Q8, correcting the values in the column, and taking mode of the values and using it to fill null values

```python

data['Q8'] = np.where(data['Q8'] == '0', 'Microsoft/Windows', data['Q8'])

mode_value = data["Q8"].mode()
data["Q8"].fillna(mode_value.values[0], inplace=True)
data['Q8']
```

```python
mode_value = data["Q9"].mode()
data["Q9"].fillna(mode_value.values[0], inplace=True)
data["Q9"]
```

```python
mode_value = data["Q10"].mode()
data["Q10"].fillna(mode_value.values[0], inplace=True)
data["Q10"]
```

```python
mode_value = data["Q11"].mode()
data["Q11"].fillna(mode_value.values[0], inplace=True)
data["Q11"]
```

```python
mode_value = data["Q12"].mode()
data["Q12"].fillna(mode_value.values[0], inplace=True)
data["Q12"]
```
